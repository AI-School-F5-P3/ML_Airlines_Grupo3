# **Tipificacion sobre los datos limpios**

```python
df_typify_clean = pd.read_csv('../data/airline_clean_data.csv')
```

```python

# Información general
df_typify_clean.info()
```

```python
df_typify_clean.columns
```

```python

#importamos el modulo de estandarizacion
from sklearn.preprocessing import StandardScaler
```

```python

# definir las columnas a normalizar
colum_normal = (['Age',
       'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes',
       ])
```

```python


print("\nRango de 'Age' antes de la normalización:")
print(df_typify_clean['Age'].min(), df_typify_clean['Age'].max())
scaler = StandardScaler()
# Aplicar el scaler
df_typify_clean[colum_normal ] = scaler.fit_transform(df_typify_clean[colum_normal ])

print("\nRango de 'Age' después de la normalización:")
print(df_typify_clean['Age'].min(), df_typify_clean['Age'].max())
```

```python

# Crear el scaler
scaler = StandardScaler()

# Aplicar el scaler a las columnas seleccionadas
df_typify_clean[colum_normal] = scaler.fit_transform(df_typify_clean[colum_normal])
```

```python


# Mostrar el DataFrame con las columnas tipificadas
print("\nDataFrame con columnas tipificadas:")
df_typify_clean
```

```python

# Guardar el DataFrame preprocesado en un archivo CSV.
df_typify_clean.to_csv('../data/airline_passenger_satisfaction_model.csv', index=False)
```
